```python
import numpy as np
import pandas as pd
from scipy.spatial import distance
from scipy.spatial import distance_matrix

from time import time
```

```python
def matprint(mat, fmt="g"):

    col_maxes = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]

    for x in mat:

        for i, y in enumerate(x):

            print(("{:"+str(col_maxes[i])+fmt+"}").format(y), end="  ")

        print("")
```

```python
import random

# Generate random points
# INPUT STUFF HERE:
numPoints = 100
dims = (100,100)
scale = 20
###

features = []

for i in range(numPoints):
    features.append( (random.randint(-dims[0],dims[0]*2), random.randint(-dims[1], dims[1]*2)) )
    
# for i in range(dims[1]):
#     features.append( (5,i) )
    
```

```python
# Brute Force Solution
t0 = time()
bruteForceSolution = np.full((dims[0], dims[1]), np.inf)
for r in range(dims[0]):
    for c in range(dims[1]):
        for feature in features:
            temp = distance.euclidean((r,c), feature) * scale
            if temp < bruteForceSolution[r, c]:
                bruteForceSolution[r, c] = temp
            
print(f"Time passed: {time() - t0}")
matprint(bruteForceSolution)
print(features)
```

```python
# RECURSIVE algorithm using Depth First Search
def check_dist(feature, curr, quadrant, first):
    #check bounds
    if(curr[0] < 0 or curr[0] >= dims[0] or curr[1] < 0 or curr[1] >= dims[1]):
        return 1

    #find dist to curr from feature
    d = distance.euclidean(feature, curr)
    if d >= dists[curr] and not first:
        return 1
    dists[curr] = d
    arr1[curr[0]][curr[1]] = d * 20
    
    
    if quadrant == 1:
        check_dist(feature, (curr[0], curr[1] - 1), quadrant, False)
        check_dist(feature, (curr[0] + 1, curr[1]), quadrant, False)
    elif quadrant == 2:
        check_dist(feature, (curr[0], curr[1] - 1), quadrant, False)
        check_dist(feature, (curr[0] - 1, curr[1]), quadrant, False)
    elif quadrant == 3:
        check_dist(feature, (curr[0], curr[1] + 1), quadrant, False)
        check_dist(feature, (curr[0] - 1, curr[1]), quadrant, False)
    elif quadrant == 4:
        check_dist(feature, (curr[0], curr[1] + 1), quadrant, False)
        check_dist(feature, (curr[0] + 1, curr[1]), quadrant, False)
    return 0

#setting up the dictionary
arr1 = np.zeros((dims[0], dims[1]))
dists = {}
for r in range(dims[0]):
    for c in range(dims[1]):
        dists[(r,c)] = np.inf
t0 = time()
for feature in features:
#     print(f"Time passed: {time() - t0}")
    check_dist(feature, feature, 1, True)
    check_dist(feature, feature, 3, True)
    check_dist(feature, (feature[0] - 1, feature[1] - 1), 2, True)
    check_dist(feature, (feature[0] + 1, feature[1] + 1), 4, True)
print(f"Total time passed: {time() - t0}")
print("Does the answer match the brute force approach: " + str(np.array_equal(bruteForceSolution, arr1)))
matprint(arr1)
```
